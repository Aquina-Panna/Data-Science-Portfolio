# ML Final Project

**Objective** Predicting heart rate to monitor stress level

**DataSet** The data comprises various attributes taken from signals measured using ECG
recorded for different individuals having different heart rates at the time the
measurement was taken. These various features contribute to the heart rate at the
given instant of time for the individual.
We have 3 csv files:
1. time_domain_features_train.csv - This file contains all time domain features of heart
rate for training data
2. frequency_domain_features_train.csv - This file contains all frequency domain
features of heart rate for training data
3. heart_rate_non_linear_features_train.csv - This file contains all non linear features of
heart rate for training data

Following is the data dictionary for the features you will come across in the files mentioned:
* MEAN_RR - Mean of RR intervals
* MEDIAN_RR - Median of RR intervals
* SDRR - Standard deviation of RR intervals
* RMSSD - Root mean square of successive RR interval differences
* SDSD - Standard deviation of successive RR interval differences
* SDRR_RMSSD - Ratio of SDRR / RMSSD
* pNN25 - Percentage of successive RR intervals that differ by more than 25 ms
* pNN50 - Percentage of successive RR intervals that differ by more than 50 ms
* KURT - Kurtosis of distribution of successive RR intervals
* SKEW - Skew of distribution of successive RR intervals
* MEAN_REL_RR - Mean of relative RR intervals
* MEDIAN_REL_RR - Median of relative RR intervals
* SDRR_REL_RR - Standard deviation of relative RR intervals
* RMSSD_REL_RR - Root mean square of successive relative RR interval differences
* SDSD_REL_RR - Standard deviation of successive relative RR interval differences
* SDRR_RMSSD_REL_RR - Ratio of SDRR/RMSSD for relative RR interval differences
* KURT_REL_RR - Kurtosis of distribution of relative RR intervals
* SKEW_REL_RR - Skewness of distribution of relative RR intervals
* uuid - Unique ID for each patient
* VLF - Absolute power of the very low frequency band (0.0033 - 0.04 Hz)
* VLF_PCT - Principal component transform of VLF
* LF - Absolute power of the low frequency band (0.04 - 0.15 Hz)
* LF_PCT - Principal component transform of LF
* LF_NU - Absolute power of the low frequency band in normal units
* HF - Absolute power of the high frequency band (0.15 - 0.4 Hz)
* HF_PCT - Principal component transform of HF
* HF_NU - Absolute power of the highest frequency band in normal units
* TP - Total power of RR intervals
* LF_HF - Ratio of LF to HF
* HF_LF - Ratio of HF to LF
* SD1 - Poincaré plot standard deviation perpendicular to the line of identity
* SD2 - Poincaré plot standard deviation along the line of identity
* Sampen - sample entropy which measures the regularity and complexity of a timeseries
* higuci - higuci fractal dimension of heartrate
* datasetId - ID of the whole dataset
* condition - condition of the patient at the time the data was recorded
* HR - Heart rate of the patient at the time of data recorded

```
## independent variable and dependent variable

X = df_all
y = target

## spliting to test and training data set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)

print("Size of X_train",X_train.shape)
print("Size of X_test",X_test.shape)
print("Size of y_train",y_train.shape)
print("Size of y_test",y_test.shape)
```

```
## Scaling the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# scaler.fit(X_train)
X_train_scaled = scaler.fit_transform(X_train)
print("scaled train data after tranformation\n",X_train_scaled)
# scaler.fit(X_test)
X_test_scaled = scaler.fit_transform(X_test)
print("scaled test data after tranformation\n",X_test_scaled)
```

```
## Feature selection using Principle componant analysis

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_train_scaled)
X_pca = pca.transform(X_train_scaled)
print("Original shape: {}".format(str(X_train_scaled.shape)))
print("Reduced shape: {}".format(str(X_pca.shape)))
print("PCA components:\n{}".format(pca.components_))

plt.matshow(pca.components_, cmap='viridis')
plt.yticks([0, 1], ["first component", "Second component"])
plt.colorbar()
plt.xticks(range(len(X_train.columns)), X_train.columns, rotation=60, ha='left')
plt.xlabel("Feature")
plt.ylabel("Principal components")
```

```
## feature select using linear regression

from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SelectFromModel

lr = LinearRegression()
select = SelectFromModel(lr, threshold='median')
select.fit(X_train, y_train)
X_Select = select.transform(X_train)
print("X_train.shape: {}".format(X_train.shape))
print("X_Select.shape: {}".format(X_Select.shape))
print("X_Select Feature:",select.estimator_.coef_)

imp_feature = pd.Series(select.estimator_.coef_, index=X_train.columns)
imp_feature.nlargest(5).plot(kind='barh')
plt.title("Important Features based on Linear Regression")
plt.show()
```

```
## Model linear regression with all feature and PCA

models_score ={}
model_test_score ={}


lr.fit(X_train_scaled,y_train)
models_score['LinearRegression'] = lr.score(X_train_scaled,y_train)
model_test_score['LinearRegression'] = lr.score(X_test_scaled,y_test)


print("Linear Regression Score with all feature: Train ", models_score['LinearRegression'],"\n Test ",model_test_score['LinearRegression'])
lr_pca = LinearRegression()
lr_pca.fit(X_pca,y_train)
models_score['LinearRegressionPCA'] = lr_pca.score(X_pca,y_train)
# model_test_score['LinearRegression'] = lr_pca.score(X_test_scaled,y_test)
print("Linear Regression score with PCA : Train",models_score['LinearRegressionPCA'],"\n Test ",)

```

```
############### Decision Tree- full

from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor(random_state=0)
dtree.fit(X_train_scaled,y_train)
models_score['FullDecisionTree'] = dtree.score(X_train_scaled,y_train)
model_test_score['FullDecisionTree'] = dtree.score(X_test_scaled,y_test)
print("Decision tree score, train: ", models_score['FullDecisionTree'],"\nTest: ",model_test_score['FullDecisionTree'])
```

```
## hyperparameter tuning using RandomSearchCV for:
## Decision Tree Regressor
## LASSO Regression
## Ridge Regression

from scipy.stats import randint 
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor(random_state=0)

dtree_param = {"max_depth":[2,4,6,8,10,12],
               "max_features": randint(1, 9),
               "min_samples_leaf": randint(1, 9)
}

from sklearn.model_selection import RandomizedSearchCV
rs_dtree = RandomizedSearchCV(dtree,dtree_param,cv=5)
rs_dtree.fit(X_train_scaled,y_train)
models_score['TunedDecisionTree'] = rs_dtree.score(X_train_scaled,y_train)
model_test_score['TunedDecisionTree'] = rs_dtree.score(X_test_scaled,y_test)
print("Tuned Dtree Score : Train: ",rs_dtree.score(X_train_scaled,y_train),"\nTest: ",model_test_score['TunedDecisionTree'])
print(rs_dtree.best_params_)
```

```
############ LASSO


from sklearn.model_selection import RandomizedSearchCV
from sklearn import linear_model

lasso = linear_model.Lasso(random_state=0)

lasso_param = {"alpha":[0.0001, 0.001, 0.005, 0.01, 0.05, 0.5, 1]
               }
rs_lasso = RandomizedSearchCV(lasso,lasso_param,cv=5,n_iter=50)
rs_lasso.fit(X_train_scaled,y_train)
models_score['Lasso'] = rs_lasso.score(X_train_scaled,y_train)
model_test_score['Lasso'] = rs_lasso.score(X_test_scaled,y_test)
print("Lasso Score : Train ",rs_lasso.score(X_train_scaled,y_train),"\nTest ",model_test_score['Lasso'])
print(rs_lasso.best_params_)
```

```
############## Ridge

from sklearn.linear_model import Ridge
ridge = Ridge(random_state=0)

ridge_param = {"alpha":[0.0001, 0.001, 0.005, 0.01, 0.05, 0.5, 1],
               "solver":['auto','saga','lsqr','svd']
               }
rs_ridge = RandomizedSearchCV(ridge,ridge_param,cv=5,n_iter=50)
rs_ridge.fit(X_train_scaled,y_train)
models_score['Ridge'] = rs_ridge.score(X_train_scaled,y_train)
model_test_score['Ridge'] = rs_ridge.score(X_test_scaled,y_test)
print("Ridge Score : Train:",rs_ridge.score(X_train_scaled,y_train),"\ntest: ",model_test_score['Ridge'])
print(rs_ridge.best_params_)
```

```
## Neural network

from sklearn.neural_network import MLPRegressor
nn = MLPRegressor(random_state=1, max_iter=500).fit(X_train_scaled, y_train)
print("Neural Network score: Train: ",nn.score(X_train_scaled,y_train), "\nTest:",nn.score(X_test_scaled,y_test))
models_score['NeuralNetwork'] = nn.score(X_train_scaled,y_train)
model_test_score['NeuralNetwork'] = nn.score(X_test_scaled,y_test)
```

```
### Gradiant Boosting 


from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(random_state = 0)
gb.fit(X_train_scaled, y_train)
models_score['GBM'] = gb.score(X_train_scaled,y_train)
model_test_score['GBM'] = gb.score(X_test_scaled,y_test)
print("GBM score: Train : ",models_score['GBM']," \nTest: ", model_test_score['GBM'])
```

```
### Computing and comparing the MSE for all the models

from sklearn.metrics import mean_squared_error

models_mse = {}

y_pred_lr = lr.predict(X_test_scaled)
models_mse['LinearRegression'] = round(mean_squared_error(y_test,y_pred_lr),3)
print("MSE for Liner Regression Model ",models_mse['LinearRegression'])

# y_pred_pca = lr_pca.predict(X_test_scaled)
# models_mse['LinearRegressionPCA']  = round(mean_squared_error(y_test,y_pred_pca),3)
# print("MSE for dtree Model ",models_mse['LinearRegressionPCA'])

dtree.fit(X_train_scaled,y_train)
y_pred_dtree = dtree.predict(X_test_scaled)
models_mse['FullDecisionTree']  = round(mean_squared_error(y_test,y_pred_dtree),3)
print("MSE for full DTree Model ",models_mse['FullDecisionTree'])

y_pred_dtreeP = rs_dtree.predict(X_test_scaled)
models_mse['TunedDecisionTree']  = round(mean_squared_error(y_test,y_pred_dtreeP),3)
print("MSE for Tune Dtree Model ",models_mse['TunedDecisionTree'])

y_pred_lasso = rs_lasso.predict(X_test_scaled)
models_mse['Lasso']  = round(mean_squared_error(y_test,y_pred_lasso),3)
print("MSE for Lasso Model ",models_mse['Lasso'])


y_pred_ridge = rs_ridge.predict(X_test_scaled)
models_mse['Ridge'] = round(mean_squared_error(y_test,y_pred_ridge),3)
print("MSE for Ridge Model ",models_mse['Ridge'])


y_pred_nn = nn.predict(X_test_scaled)
models_mse['NeuralNetwork'] = round(mean_squared_error(y_test,y_pred_nn),3)
print("MSE for Neural Network ",models_mse['NeuralNetwork'])


y_pred_gb = gb.predict(X_test_scaled)
models_mse['GBM'] = round(mean_squared_error(y_test,y_pred_gb),3)
print("MSE for GBM ",models_mse['GBM'])
```
